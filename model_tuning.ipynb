```python
from Qlearning import QAgent
from TestEnv import HydroElectric_Test
import matplotlib.pyplot as plt
import numpy as np 
import random
import optuna
```

```python
seed = 5
np.random.seed(seed)
random.seed(seed)
```

```python
env = HydroElectric_Test("train.xlsx")
env.reset(seed=seed)
agent = QAgent(env)
```

```python
parameter_tuning_dict = {
    'learning_rate': (0.01, 1.0),
    'epsilon': (0.01, 1.0)
}
agent.hyperparameter_tuning(parameter_tuning_dict)
```

```python
optuna.visualization.plot_optimization_history(agent.study)
```

```python
optuna.visualization.plot_param_importances(agent.study)
```

```python
optuna.visualization.plot_contour(agent.study, params=['learning_rate', 'epsilon'])
```
